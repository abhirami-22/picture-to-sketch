<h2> Image processing using Python OpenCV </h2><br>
Generating sketch from image using Python 

```python
# importing libraries
import cv2
import matplotlib.pyplot as plt
```

```python
# reading image using opencv
img = cv2.imread("obama.jpg")
```

```python
# displaying the image in new window
cv2.imshow("Obama",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# using pyplot to display the image in the Jupyter notebook itself
# image in BGR
plt.imshow(img)
plt.axis(False)
plt.show()
```

```python
# Converting BGR to RGB 
# Method 1 
plt.imshow(img[:,:,::-1])
plt.axis(False)
plt.show()
```

```python
# Converting BGR to RGB
# Method 2 - Using cvtColor method of OpenCV
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.show()
```

Converting a photo to pencil sketch involves the following steps:<br>
Step 1: Convert to Grey Image<br>
Step 2: Invert Image<br>
Step 3: Blur image<br>
Step 4: Invert Blurred Image<br>
Step 5: Sketch<br>
Step 6: Save Sketch<br>

```python
# Step 1: Convert to gray mage
gray_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)    # in BGR
gray = cv2.cvtColor(gray_img, cv2.COLOR_GRAY2RGB) # converting to RGB
plt.imshow(gray)
plt.axis(False)
plt.show()
```

```python
# Step 2: Invert image
invert_img=cv2.bitwise_not(gray_img)                 # in BGR
invert = cv2.cvtColor(invert_img, cv2.COLOR_BGR2RGB) # converting to RGB
plt.imshow(invert)
plt.axis(False)
plt.show()
```

```python
# Step 3: Blur image
blur_img=cv2.GaussianBlur(invert_img, (111,111),0) # in BGR
blur = cv2.cvtColor(blur_img, cv2.COLOR_GRAY2RGB)  # converting to RGB
plt.imshow(blur)
plt.axis(False)
plt.show()
```

```python
# Step 4: Invert blurred image
invblur_img=cv2.bitwise_not(blur_img) # invblur image in BGR
invblur = cv2.cvtColor(invblur_img, cv2.COLOR_GRAY2RGB) # invblur image in RGB
plt.imshow(invblur)
plt.axis(False)
plt.show()
```

```python
# Step 5: Sketch
sketch_img=cv2.divide(gray_img,invblur_img, scale=256.0) # sketch = gray image / inverted blurred image
sketch = cv2.cvtColor(sketch_img, cv2.COLOR_GRAY2RGB)    # sketch image in RGB
plt.imshow(sketch)
plt.axis(False)
plt.show()
```

```python
# Step 6: Save Sketch
cv2.imwrite('sketch.jpg', sketch_img)
```
